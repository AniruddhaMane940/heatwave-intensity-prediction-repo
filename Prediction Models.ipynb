# HyperParameter Tunning For ANN

```python
from sklearn.model_selection import GridSearchCV
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier 
```

```python
def getModel(neurons1,neurons2):
    model = Sequential()
    model.add(Dense(neurons1, input_dim=12, activation='relu',input_shape = (4,)))
    model.add(Dense(neurons2, activation='tanh'))
    model.add(Dense(1,activation='linear'))
    model.compile(loss='mean_squared_error', optimizer='Adam')
    return model
```

```python
neurons1=[1024,512,256,128]
neurons2=[1024,512,256,128]
epochs = [50,60,70,80]
```

```python
param_grid = dict(epochs=epochs, neurons1=neurons1,neurons2=neurons2)
```

```python
Kmodel = KerasClassifier(build_fn=getModel, verbose=1)
grid = GridSearchCV(estimator=Kmodel, param_grid=param_grid, scoring='neg_mean_absolute_percentage_error')
grid_result = grid.fit(x_train, y_train)
```

```python
grid_result.best_params_
```

```python
grid_result.best_score_
```
